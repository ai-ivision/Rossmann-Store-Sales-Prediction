## *Rossmann Store Sales Prediction*

> *Problem Statement :*

>> *Challenge:*

Rossmann, a drug store chain operating in 7 European countries, currently relies on individual store managers to predict daily sales for up to six weeks in advance. This approach leads to inconsistent accuracy due to varying levels of experience and information available to each manager.

>> *Objective:*

Develop a machine learning model that can accurately forecast daily sales for Rossmann stores across different locations and time periods.

>> *Data:*

The model will be trained on a historical dataset containing sales information for over 1,100 Rossmann stores. This data is expected to include features such as:

**Sales:** Daily sales figures <br>
**Store:** Information about the store location and type <br>
**Promotions:** Ongoing promotions and discounts offered <br>
**Competition:** Presence and proximity of competing stores <br>
**Holidays:** National and school holidays <br>
**Seasonality:** Day of the week, week of the year, month, etc.

```python
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

merged_df.dtypes
```

```python
merged_df['Year'] = merged_df.Date.dt.year

merged_df['Month'] = merged_df.Date.dt.month

merged_df['Day'] = merged_df.Date.dt.day

merged_df['Week'] = merged_df.Date.dt.weekday

merged_df.head()
```

```python
# No od months that competition has existed for
merged_df['MonthsCompetitionOpen'] =  12 * (merged_df['Year'] - merged_df['CompetitionOpenSinceYear']) + (merged_df['Month'] - merged_df['CompetitionOpenSinceMonth'])

merged_df.loc[merged_df['CompetitionOpenSinceYear'] == 0, 'MonthsComprtitionOpen'] = 0
```

```python
# No of weeks that promotion has existed for
merged_df['WeeksPromoOpen'] = 12 * (merged_df['Year'] - merged_df['Promo2SinceYear']) + (merged_df['Date'].dt.weekday - merged_df['Week'])

merged_df.loc[merged_df['Promo2SinceYear'] == 0, 'WeeksPromoOpen'] = 0
```

```python
x = [
    'Store',
    'Customers',
    'CompetitionDistance',

    'Promo', 
    'Promo2',

    'CompetitionOpenSinceMonth', 
    'CompetitionOpenSinceYear', 
    'Promo2SinceWeek',
    'Promo2SinceYear',

    'StateHoliday',
    'StoreType',
    'Assortment',

    'AvgSales',
    'AvgCustomers', 
    'AvgSalesPerCustomer',

    'MedSales',
    'MedCustomers',
    'MedSalesPerCustomer',

    'DayOfWeek',
    'Week',
    'Day', 
    'Month',
    'Year',
]

X_ = merged_df[x]
Y_ = np.log(merged_df['Sales'])

X_train, X_test, Y_train, Y_test = train_test_split(X_, Y_, test_size=0.2, random_state=10)
```

```python

import sklearn
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
```

```python
import sklearn.metrics


sorted(sklearn.metrics.SCORERS.keys())
```

```python
params = {
    'n_estimators' : [100, 500, 1000, 1500],
    'max_depth' : [ 2, 4, 6, 8]
}

xgboost_tree = xgb.XGBRegressor(
    eta = 0.1,
    min_child_weight = 2,
    subsample = 0.8,
    colsample_bytree = 0.8,
    tree_method='exact',
    reg_alpha = 0.05,
    silent = 0,
    random_state=1023
)

grid_search = GridSearchCV(
    estimator=xgboost_tree,
    param_grid=params,
    cv=5,
    n_jobs=-1,
    verbose=1,
    scoring='neg_mean_squared_error',
)

grid_res = grid_search.fit(X_train, Y_train)

grid_score = grid_search.score(X_train, Y_train)
best_params = grid_search.best_params_


print(f"Grid Score    :   {grid_score}")
print(f"Best Params     :   {best_params}")
```

```python
pred = grid_search.predict(X_test)
```

```python
from math import sqrt

rmse = sqrt(mean_squared_error(np.exp(Y_test), np.exp(pred)))

print(f"The Root MEAN Squared error is      :      {rmse}")
```
